```python
from matplotlib import pyplot as plt
from math import sin
from math import cos
from math import pow
```

```python
print(cos(1))
```

```python
def f(x):
    list_y = []
    for item in x:
        y = -12 * pow(item,4) * sin(cos(item)) - 18 * pow(item,3) + 5 * pow(item,2) + 10 * item - 30
        list_y.append(round(y, 3))
    return list_y
```

```python
def x():
    count = -100
    list_x = []
    while count <= 100:
        list_x.append(round(count, 3))
        count += 0.2
    return list_x
```

```python
x_arr = x()
y_arr = f(x_arr)
plt.plot(x_arr, y_arr)
```
